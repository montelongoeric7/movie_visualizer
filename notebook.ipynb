```python
def get_movies_by_genre(genre):
  
  url = f'https://www.imdb.com/search/title/?genres={genre}&view=simple'
  headers = {"User-Agent":"Mozilla/5.0 (Macintosh; Intel Mac OS X 10.15; rv:89.0) Gecko/20100101 Firefox/89.0"}
  response_object = requests.get(url,headers=headers)
  print(response_object.status_code)
  decoded_content = response_object.content.decode("UTF-8")
  soup = BeautifulSoup(decoded_content, "html.parser")

  answer_ids = []
  for title in soup.find_all(attrs={'class':'col-title'}):
    answer_ids.append(title.a.get('href').split('/')[2])
  

  return answer_ids
```

```python
def get_omdb_movie_details(id):
  my_key = '4e4303e6'
 
  url_omdb = f'http://www.omdbapi.com/?apikey={my_key}&i={id}'
  response = requests.get(url_omdb).json()
  del response['Ratings']

  return response

```

```python
def get_movie_details_by_genere(genre):
  top_movies = get_movies_by_genre(genre)
  movie_details = []

  for id in top_movies:
    movie_details.append(get_omdb_movie_details(id))
  df = pd.DataFrame(movie_details)
  return df
```

```python
comedy = get_movie_details_by_genere('comedy')
```

```python
horror = get_movie_details_by_genere('horror')
```

```python
all_movies = pd.concat([horror,comedy],axis=0)
```

```python

nlp = spacy.load('en_core_web_md')
```

```python
all_movies['embedding'] = all_movies['Plot'].apply(lambda x:nlp(x).vector)
```

```python
from sklearn.manifold import TSNE
```

```python
tsne = TSNE(n_components=2,perplexity=5) #The perplexity is default=30, has to do with size of the neighborhoods, 2.)iterations: if we dont give it enough time it might not find a good,(Default is 1000) 
```

```python
import numpy as np
```

```python
two_d_embedding = tsne.fit_transform(np.stack(all_movies['embedding']))
```

```python
all_movies['2DX'] = two_d_embedding[:,0]
all_movies['2DY'] = two_d_embedding[:,1]
```
